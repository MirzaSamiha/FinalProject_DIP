```python
import cv2
import os
import matplotlib.pyplot as plt
import numpy as np

# Function to calculate PSNR
def calculate_psnr(original, noisy):
    if original.shape != noisy.shape:
        noisy = cv2.cvtColor(noisy, cv2.COLOR_GRAY2RGB)

    mse = np.mean((original - noisy)**2)
    psnr = 20 * np.log10(255 / np.sqrt(mse))
    return psnr

# Function to perform edge detection and save output images
def perform_edge_detection_and_save(input_folder, output_folder, edge_method, title):
    os.makedirs(output_folder, exist_ok=True)

    for filename in os.listdir(input_folder):
        
        input_path = os.path.join(input_folder, filename)

        # Read the image
        image = cv2.imread(input_path)

        # Apply edge detection method
        edges = edge_method(image)

        # Save the output image
        output_path = os.path.join(output_folder, f'{title}_{filename}')
        cv2.imwrite(output_path, edges)

# Define paths to your dataset folder and the output folder
dataset_folder = r'C:\Users\samih\Documents\dip\dip_project\daisy'
output_parent_folder = r'C:\Users\samih\Documents\dip\dip_project\edm1'


edge_methods = [
    ('Sobel', sobel_edge_detection),
    ('Prewitt', prewitt_edge_detection),
    ('Canny', canny_edge_detection),
    
]

# Perform edge detection and save output images for each method
for title, method in edge_methods:
    output_folder = os.path.join(output_parent_folder, title)
    perform_edge_detection_and_save(dataset_folder, output_folder, method, title)
```

```python
import os
import cv2
import numpy as np

def process_image(file_path, output_folder):
    # Read the image
    image = cv2.imread(file_path, cv2.IMREAD_GRAYSCALE)

    # Binarization (assuming you've done this step, adjust if needed)
    _, binary_image = cv2.threshold(image, 128, 255, cv2.THRESH_BINARY)

    # Complement the binary image
    complemented_image = cv2.bitwise_not(binary_image)

    # Convert the image to 3 channels for watershed
    image_color = cv2.cvtColor(image, cv2.COLOR_GRAY2BGR)

    # Watershed Transform
    _, markers = cv2.connectedComponents(binary_image)
    markers = markers + 1
    markers[complemented_image == 255] = 0
    watershed_result = cv2.watershed(image_color, markers)

    # Extract watershed ridge pixels (0-valued pixels)
    watershed_ridge_pixels = np.uint8(watershed_result == -1)

    # Binary AND of original binary image and complement of watershed ridge pixels
    segmented_result = cv2.bitwise_and(binary_image, cv2.bitwise_not(watershed_ridge_pixels))

    # Debugging information
    print(f"Processing: {file_path}")

    # Create output folders if they don't exist
    os.makedirs(output_folder, exist_ok=True)

    # Save the final segmented result
    cv2.imwrite(os.path.join(output_folder, f'{os.path.basename(file_path)}_segmented_result.png'), segmented_result)

if __name__ == "__main__":
    dataset_path = r'C:\Users\samih\Documents\dip\dip_project\Leaveso'
    output_path = r'C:\Users\samih\Documents\dip\dip_project\outputsegmleaves'

    # Process each image in the dataset
    for file_name in os.listdir(dataset_path):
        if file_name.endswith(".png") or file_name.endswith(".jpg"):
            file_path = os.path.join(dataset_path, file_name)
            process_image(file_path, output_path)
```
